# Softmax

```python
v = np.random.randn(4)

v_exp = np.exp(v)
soma = np.sum(v_exp)
resultado = v_exp/soma
print(v)
print(resultado)
```

```python
softfun = torch.nn.Softmax(dim=0)
softfun
```

```python
res = softfun(torch.Tensor(v))
res
```

## T-Test

```python
import scipy.stats as stats
```

```python
num1 = 40
num2 = 45
mu1 = 1
mu2 = 1.11

data1 = mu1 + np.random.rand(num1)
data2 = mu2 + np.random.rand(num2)
```

```python
t,p = stats.ttest_ind(data2,data1)
print(t)
print(p)
```

```python
import matplotlib.pyplot as plt

plt.plot(np.zeros(num1) + np.random.rand(num1)/10, data1, "bo", markerfacecolor="w")
plt.plot(np.ones(num2) + np.random.rand(num2)/10, data2, "rx")
```
