```python
nv1 = np.array([1,2,3,4])
nv2 = np.array([0,1,0,-1])

print(np.dot(nv1,nv2))


print(np.sum(nv1*nv2))
```

# Softmax

```python
v = np.random.randn(4)

v_exp = np.exp(v)
soma = np.sum(v_exp)
resultado = v_exp/soma
print(v)
print(resultado)
```

```python
softfun = torch.nn.Softmax(dim=0)
softfun
```

```python
res = softfun(torch.Tensor(v))
res
```
