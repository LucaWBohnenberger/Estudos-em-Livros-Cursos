```python
from sklearn.datasets import load_iris

X = load_iris()
```

```python
y = X.target
X = X.data
```

```python
X
```

```python
import numpy as np
X_bias = np.c_[np.ones([len(X), 1]), X]
```

```python
X_bias
```

```python
test_ratio = 0.2
validation_ratio = 0.2
total_size = len(X_bias)

test_size = int(total_size * test_ratio)
validation_size = int(total_size * validation_ratio)
train_size = total_size - test_size - validation_size

rnd_indices = np.random.permutation(total_size)

X_train = X_bias[rnd_indices[:train_size]]
y_train = y[rnd_indices[:train_size]]
X_valid = X_bias[rnd_indices[train_size:-test_size]]
y_valid = y[rnd_indices[train_size:-test_size]]
X_test = X_bias[rnd_indices[-test_size:]]
y_test = y[rnd_indices[-test_size:]]
```

```python
def one_hot(vetor):
    colunas = vetor.max() + 1
    linhas = len(vetor)
    y_one_hot = np.zeros((linhas, colunas))
    for i in range(linhas):
        y_one_hot[i, vetor[i]] = 1
    return y_one_hot

one_hot(y_train[:10])
```

```python
y_train_one = one_hot(y_train)
y_valid_one = one_hot(y_valid)
y_test_one = one_hot(y_test)
```

```python
def softmax(logits):
    exps = np.exp(logits)
    exp_sums = np.sum(exps, axis=1, keepdims=True)
    return exps / exp_sums
    
```

```python
n_inputs = X_train.shape[1]
n_outputs = len(np.unique(y_train))
print(n_inputs, n_outputs)
```

```python
Theta = np.random.randn(n_inputs, n_outputs)
eta = 0.01

logits = X_train.dot(Theta)
y_proba = softmax(logits)
y_proba
```

```python
error = y_proba - y_train_one
error
```

```python
gradients = 1/len(X_train) * X_train.T.dot(error)

Theta = Theta - eta * gradients
```

```python
epochs = 40000
Theta = np.random.randn(n_inputs, n_outputs)
eta = 0.01
epsilon = 1e-7
print_interval = 10000

for epoch in range(epochs):
    scores = X_train.dot(Theta)
    
    y_proba = softmax(scores)
    error = y_proba - y_train_one
    
    gradients = 1/len(X_train) * X_train.T.dot(error)
    
    if (epoch % print_interval) == 0:
        loss = -np.mean(np.sum(y_train_one * np.log(y_proba + epsilon), axis=1))
        print("Epoch: ", epoch, " Loss: ", loss)
    
    Theta -= gradients * eta
    
```

```python
val = X_train[0]
print(val, y_train[0])
```

```python
y_score = X_test.dot(Theta)
y_pred = np.argmax(y_score, axis=1)

from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
