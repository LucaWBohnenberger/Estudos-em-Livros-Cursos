```python
from sklearn.datasets import load_iris

data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
import numpy as np
```

```python
X_bias = np.c_[np.ones(X.shape[0]), X]
```

```python
X_bias
```

```python
test_rate = 0.3
total_size = len(X)

rnd_indice = np.random.permutation(total_size)

test_size = int(total_size * test_rate)

X_train = X_bias[rnd_indice[:-test_size]]
y_train = y[rnd_indice[:-test_size]]
X_test = X_bias[rnd_indice[-test_size:]]
y_test = y[rnd_indice[-test_size:]]
```

```python
def OneHotEncoding(vetor):
    linhas = len(vetor)
    colunas = np.max(vetor) + 1
    matriz = np.zeros([linhas,colunas])
    for i in range(linhas):
        matriz[i][vetor[i]] = 1
    return matriz
```

```python
y_train_one = OneHotEncoding(y_train)
y_test_one = OneHotEncoding(y_test)
```

```python
X_train
```

```python
def softmax(logits):
    exps = np.exp(logits)
    soma = np.sum(exps, axis=1, keepdims=True)
    return exps/soma
```

```python
linhas = X_train.shape[1]
colunas = np.max(y_train) + 1
theta = np.random.randn(linhas, colunas)
```

```python
eta = 0.001
epochs = 50000
for epoch in range(epochs):    
    # Calculo do score e probabilidade
    scores = X_train.dot(theta)
    y_prob = softmax(scores)

    erro = y_prob - y_train_one

    #treinamento
    gradient = 1/len(X_train) * X_train.T.dot(erro)
    theta -= eta*gradient
```

```python
# Previs√£o
score = X_test.dot(theta)
pred = np.argmax(score, axis=1)

# Metricas
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```
