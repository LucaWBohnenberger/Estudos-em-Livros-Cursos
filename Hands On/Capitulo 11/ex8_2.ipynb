```python
from tensorflow import keras
```

```python
(X_train,y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
model = keras.models.Sequential([
    keras.layers.Normalization(-1),
    keras.layers.Flatten(input_shape=[32,32,3]),
])
for i in range(20):
    model.add(keras.layers.Dense(100, activation="selu", kernel_initializer="lecun_normal")),
    
model.add(keras.layers.AlphaDropout(0.1))

norm = model.layers[0]
norm.adapt(X_train)
    
model.add(keras.layers.Dense(10, activation="softmax"))
```

```python
model.compile(loss=keras.losses.SparseCategoricalCrossentropy(), optimizer=keras.optimizers.Nadam(learning_rate=0.0005), metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, validation_split=0.2, epochs=150, 
                    callbacks=[keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)])

```

```python
model.save("model4.keras")
```

```python
model.evaluate(X_test, y_test)
```

```python
import numpy as np
y_probas = np.stack([model(X_test, training=True) for sample in range(100)])
y_proba = y_probas.mean(axis=0)
from sklearn.metrics import accuracy_score
y_pred = np.argmax(y_proba, axis=1)
accuracy_score(y_test, y_pred)
```
