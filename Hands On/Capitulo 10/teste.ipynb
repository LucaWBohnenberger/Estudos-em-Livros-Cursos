```python
from tensorflow import keras
```

```python
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
X = data.data
y = data.target
```

```python
X
```

```python
_input = keras.layers.Input(shape=X.shape[1:])
hidden = keras.layers.Dense(30, activation="relu")(_input)
hidden2 = keras.layers.Dense(30, activation="relu")(hidden)
concat = keras.layers.Concatenate(1)([_input, hidden2])
output = keras.layers.Dense(1)(concat)
model = keras.Model(_input, output)

model.compile(loss="mse", optimizer=keras.optimizers.SGD(0.0000001))
```

```python
early_stopping = keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)
history = model.fit(X, y, epochs=250, validation_split=0.2, callbacks=[early_stopping])
```
